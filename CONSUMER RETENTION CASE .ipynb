REMOVING OUTLIERS

from the dataset using its exact position in the dataset because in all the above methods of detecting the outliers end result is the list of all those data items that satisfy the outlier definition according to the method used.

```python
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
import seaborn as sns

df=pd.read_csv('C:/Users/user/Desktop/DT Python/customer_retention_dataset.csv')

# IQR
Q1 = np.percentile(df['4 What is the Pin Code of where you shop online from?'], 25, interpolation = 'midpoint')

Q3 = np.percentile(df['4 What is the Pin Code of where you shop online from?'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", df.shape)

# Upper bound
upper = np.where(df['4 What is the Pin Code of where you shop online from?'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(df['4 What is the Pin Code of where you shop online from?'] <= (Q1-1.5*IQR))

# Removing the Outliers
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

print("New Shape: ", df.shape)

sns.boxplot(x='4 What is the Pin Code of where you shop online from?', data=df)
```
